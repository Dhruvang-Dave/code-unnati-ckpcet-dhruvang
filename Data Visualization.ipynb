```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
#x axis 
x = [10,20,30,40,50,60,70]
#y axis
y = [15,55,36,12,20,40,56]

# x label
plt.xlabel("Sales")
# y label
plt.ylabel("Profit ($-Dollar)")
#title
plt.title("About 2021 - 2022 Sales")
#ploting the points
plt.plot(x,y,linestyle="dashed",color='purple',marker='o', ms=10)
plt.bar(x,y)
plt.show()
```

```python
#subplot
#plot 1
x = np.array([1,2,3,4,5])
y = np.array([6,7,8,9,10])

#plot 2
z = np.array([1,2,3,4,5])
w = np.array([10,17,18,19,20])

#plot 3
p = np.array([1,2,3,4,5])
q = np.array([16,17,18,19,10])

#plot 
r = np.array([11,12,13,14,15])
s = np.array([10,17,18,19,20])

plt.subplot(1,2,2)
plt.plot(z,w)
plt.subplot(1,2,1) #no of row, no of column , last one is position of graph
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(z,w)
plt.subplot(2,2,2) #no of row, no of column , last one is position of graph
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(p,q)
plt.subplot(2,2,4) #no of row, no of column , last one is position of graph
plt.plot(r,s)
```
